```python
def winning_score(x,y):

    # How many matched played by 2 team against each other
    against_df = ipl[((ipl["Team1"] == x) & (ipl["Team2"] == y))|((ipl["Team1"] == y) & (ipl["Team2"] == x))]
    x_winner = against_df[against_df["WinningTeam"] == x].shape[0]
    y_winner = against_df[against_df["WinningTeam"] == y].shape[0]
    return f"Match win by team {x} is {x_winner} and Match win by {y} is {y_winner}"

winning_score("Rajasthan Royals","Gujarat Titans")
winning_score("Kolkata Knight Riders","Royal Challengers Bangalore")
```

```python
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()

genres.apply(foo)
```

```python
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group

genres.apply(rank_movie)
```

```python
def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```
